```python
import pandas as pd
import seaborn as sns
from scipy.stats import levene, bartlett
```

```python
df = pd.read_csv('/content/drive/MyDrive/Assignment_Statistics/data.csv')
```

```python
# Check whether the distribution of all the classes are the same or not.

sns.kdeplot(df[' Blood Pressure Before (mmHg)'])
```

```python
sns.kdeplot(df[' Blood Pressure After (mmHg)'])
```

```python
# Check for the equality of variance/

levene_test = levene(df[' Blood Pressure Before (mmHg)'], df[' Blood Pressure After (mmHg)'])

bartlett_test = bartlett(df[' Blood Pressure Before (mmHg)'], df[' Blood Pressure After (mmHg)'])

print("Levene's test p-value:", levene_test.pvalue)
print("Bartlett's test p-value:", bartlett_test.pvalue)
```

```python
# Check the equality of mean for between all the classes.
from scipy.stats import f_oneway

f_statistic, p_value = f_oneway(df[' Blood Pressure Before (mmHg)'], df[' Blood Pressure After (mmHg)'])

print("F-statistic:", f_statistic)
print("p-value:", p_value)
```

```python
import pandas as pd
import numpy as np
import statistics
import matplotlib.pyplot as plt
import scipy.stats as stats
from scipy.stats import pearsonr
```

```python
df = pd.read_csv('/content/drive/MyDrive/Assignment_Statistics/data.csv')
```

```python
df.head()
```

```python
# Measure the dispersion in both and interpret the results.
print("Range of Blood Pressure After (mmHg) is : ", df[' Blood Pressure After (mmHg)'].max() - df[' Blood Pressure After (mmHg)'].min())
print("Range of Blood Pressure Before (mmHg) is : ", df[' Blood Pressure After (mmHg)'].max() - df[' Blood Pressure Before (mmHg)'].min())
```

```python
print("Variance of Blood Pressure After (mmHg) is : ", statistics.variance(df[' Blood Pressure After (mmHg)']))
print("Standard Deviation of Blood Pressure After (mmHg) is : ", statistics.stdev(df[' Blood Pressure After (mmHg)']))
```

```python
print("Variance of Blood Pressure Before (mmHg) is : ", statistics.variance(df[' Blood Pressure Before (mmHg)']))
print("Standard Deviation of Blood Pressure Before (mmHg) is : ", statistics.stdev(df[' Blood Pressure Before (mmHg)']))
```

```python
print("IQR of Blood Pressure After (mmHg) is : ", np.percentile(df[' Blood Pressure After (mmHg)'], 75) - np.percentile(df[' Blood Pressure After (mmHg)'], 25))
print("IQR of Blood Pressure Before (mmHg) is : ", np.percentile(df[' Blood Pressure Before (mmHg)'], 75) - np.percentile(df[' Blood Pressure Before (mmHg)'], 25))
```

```python
# Calculate mean and 5% confidence interval and plot it in a graph
confidence_interval = stats.t.interval(0.95, len(df[' Blood Pressure After (mmHg)'])-1, loc=np.mean(df[' Blood Pressure After (mmHg)']), scale=stats.sem(df[' Blood Pressure After (mmHg)']))
mean = np.mean(df[' Blood Pressure After (mmHg)'])
print("Mean:", mean)
print("Confidence Interval:", confidence_interval)
```

```python
plt.hist(df[' Blood Pressure After (mmHg)'], bins=20, density=True, alpha=0.5, label='Blood Pressure After (mmHg)')
plt.axvline(x=mean, color='r', linestyle='--', label='Mean')
plt.axvline(x=confidence_interval[0], color='g', linestyle='--', label='Confidence Interval')
plt.axvline(x=confidence_interval[1], color='g', linestyle='--')
plt.legend()
plt.xlabel('Value')
plt.ylabel('Density')
plt.title('Dataset with Mean and Confidence Interval')
plt.show()
```

```python
# Calculate the Mean absolute deviation and Standard deviation and interpret the results
mad = np.mean(np.abs(np.array(df[' Blood Pressure After (mmHg)']) - np.mean(df[' Blood Pressure After (mmHg)'])))
mean = np.mean(df[' Blood Pressure After (mmHg)'])
print(mad)
print(mean)
```

```python
# Calculate the correlation coefficient and check the significance of it at 1% level of significance.
corr_coef, p_value = pearsonr(df[' Blood Pressure Before (mmHg)'], df[' Blood Pressure After (mmHg)'])
print("Correlation Coefficient:", corr_coef)

if p_value < 0.01:
    print("The correlation coefficient is significant at the 1% level.")
else:
    print("The correlation coefficient is not significant at the 1% level.")
```
